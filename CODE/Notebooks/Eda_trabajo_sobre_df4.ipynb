```python

import pandas as pd
import numpy as np


df4 = pd.read_csv ('AGUA/FAOSTAT/Codigos de productos.csv', sep= ',')
df4.head()
```

```python
df4 = df4.rename(columns={'C?digo Grupo de Productos': 'Código Grupo de Productos', 'C?digo Producto' : 'Código Producto'})
df4.head(200)
```

```python
df4[df4['Producto'].isin(['Zanahorias y nabos'])]


```

```python
df4['Grupo de Productos'].unique()
```

```python
df4 = df4.replace('C?tricos, total', 'Cítricos, total')
df4 = df4.replace('Cultivos Az?car en Bruto', 'Cultivos Azúcar en Bruto')
df4 = df4.replace('Ra?ces&Tub?rculos, total', 'Raíces&Tubérculos, total')
df4 = df4.replace('Ganader?a primaria', 'Ganadería primaria')
df4 = df4.replace('Vacas y b?falos', 'Vacas y búfalos')
df4 = df4.replace('Carne Vacuna y de B?falo', 'Carne Vacuna y de Búfalo')
df4 = df4.replace('Ganader?a procesada', 'Ganadería procesada')
df4['Grupo de Productos'].unique()
```

```python
index = df4[df4['Grupo de Productos'].isin(['Cultivos Primarios'])].index
df4[df4['Grupo de Productos'].isin(['Cultivos Primarios'])]
```

```python
# sacamos los valores unicos de Codigo Grupo de producstos
codigo_grupo_producto = df4['Código Grupo de Productos'].unique().tolist()
# nos quedamos solo con los codigos de codigo gurpo de productos cuya longitud sea 4
codigo_grupo_producto = [x for x in codigo_grupo_producto if len(x)>3]
codigo_grupo_producto
```

```python

# con esta mascara sacamos todos los codigo grupo de producto en el que aparezca las claves de la lista codigo producto
mask = df4['Código Grupo de Productos'].isin(codigo_grupo_producto)
df4[mask]# aqui lo mostramos
```

```python
#para asegurarnos que la mascara anterior contiene un valor para cada producto hacemos esta comprobación
df4.loc[mask, : ].groupby('Código Grupo de Productos')['Grupo de Productos'].nunique().sort_values(ascending=False)
```

```python
# ahora lo que estamos haciendo es el diccionario para poder reemplazar los valores en la columna Codigo Grupo de Producto

diccionario = df4.loc[mask, : ].set_index('Grupo de Productos')['Código Grupo de Productos'].to_dict()# este es el diccionario para normalizar la columna Co
```

```python
diccionario
```

```python
#sustituimos la columna Codigo Grupo de productos por la columna grupo de producto mapeada al diccionario(en el diccionario viene cada grupo de producto con su clave que es lo que ponemos
# en la columna coidog grupo de productos)

df4['Código Grupo de Productos'] = df4['Grupo de Productos'].replace(diccionario)
df4['Código Grupo de Productos'].unique()# se han asigando los valores de ganaderia primera, cultivos derivados etc, porque el valor de la la lista era inferior a 3(tienen codigo QC)
```

```python
df4.drop (['Coeficiente','C?digo HS','C?digo HS07','C?digo HS12','C?digo CPC'],axis=1, inplace= True)
df4
```

```python
df4.to_csv('AGUA/FAOSTAT/Codigos de productos_normalizado.csv', sep= ';', index= False)
```
