```python

import pandas as pd
import numpy as np


df3 = pd.read_excel ('AGUA/FAOSTAT/Cultivos y productos de ganaderia.xls')
df3.head()
df3.shape
```

```python
df4 = pd.read_csv ('AGUA/FAOSTAT/Codigos de productos_normalizado.csv', sep= ';', index_col= False )
df4
```

```python
# HAGO LA UNION DE LOS DOS DATAFRAME (EL MAESTRO Y EL QUE VOY A MANEJAR)

df_union = pd.merge(df3,df4, how= 'outer')
df_union
```

```python
df_union.info()
```

```python
#ELIMINACION DE COLUMNAS QUE NO ME INTERESAN

df_union = df_union.drop(['C?digo Elemento','C?digo ?mbito','?mbito','C?digo ?rea (FAO)', '?rea', 'C?digo a?o'], axis=1)
df_union
```

```python
# RENOMBRO LAS COLUMNAS QUE TIENEN ?


df_union= df_union.rename(columns={'A?o' : 'Año', 'C?digo Producto (FAO)': 'Código Producto_origi'})
df_union.head()
```

```python
df_union['Elemento'].unique()
```

```python

# 1 . PRIMERO ELIMINO LOS REGISTROS QUE NO ME INTERESAN

df_union = df_union.drop(df_union[df_union['Elemento'].isin(['Rendimiento','Rendimiento/Peso en carcasa','Ponedoras','Reses lecheras','?rea cosechada'])].index)
df_union['Elemento'].unique()
```

```python
# 2 . SUSTITUYO AHORA LAS ? DE LA COLUMNA ELEMENTO

valorcorr1_elemen, valorcorr2_elemen= 'Producción', 'Animales en producción/Sacrificados'

df_union = df_union.replace('Producci?n', valorcorr1_elemen)
df_union = df_union.replace('Animales en producci?n/Sacrificados', valorcorr2_elemen)

df_union['Elemento'].unique()
```

```python
df_union.info()
```

```python
# PRIMERO BUSCO LOS VALORES NAN EN CODIGO PRODUCTO_ORIGIN,(QUE ES EL CODIGO PRODUCTO DE LA TABLA ORIGINAL) Y COMPRUEBO QUE LOS VALORE DEL RESTO DE COLUMNAS
#DE LA TABLA ORIGINAL SEAN TB NAN YA QUE ESTO SIGNIFICA QUE SE HAN GENERADO POR LA UNION
#ELIMINO LAS FILAS DONDE LOS VALORES DE LA TABLA ORGINAL SON NAN PORQUE AHI NO TENGO INFORMACION, SE HAN GENERADO DE LA UNION

index =df_union.loc[df_union['Código Producto_origi'].isna(),['Elemento','Código Producto_origi','Producto','Año','Unidad','Valor', 'Código Grupo de Productos', 'Grupo de Productos','Código Producto']].index

df_union = df_union.drop(index)
df_union.info()
```

```python
#AHORA BUSCO LOS VALORES NAN EN LA COLUMNA VALOR Y MUESTRO TODAS LAS COLUMNAS PARA COMPROBAR EL ESTADO Y BORRO

index= df_union.loc[df_union['Valor'].isna(),['Elemento','Código Producto_origi','Producto','Año','Unidad','Valor', 'Código Grupo de Productos', 'Grupo de Productos','Código Producto']].index
df_union = df_union.drop(index)
df_union.info()
```

```python
df_union['Elemento'].unique()
```

```python
#AQUI COMPRUEBO, EN LA COLUMNA ELEMENTO QUE LOS VALORES Y BORRO VALORES QUE SON SUMATORIO

pd.set_option('display.max_columns', 1000)

index = df_union[df_union['Grupo de Productos'].isin(['Carne, total'])].index
df_union = df_union.drop(index)
```

```python
#ELIMINO ESTOS VALORES PORQUE NO ME INTERESAN PARA EL ANALISIS
index = df_union[df_union['Grupo de Productos'].isin(['Huevos Naturales'])].index
df_union = df_union.drop(index)
```

```python
df_union['Grupo de Productos'].unique()
```

```python
df_union['Grupo de Productos'].unique()
```

```python
# 4. SUSTITUYO AHORA LAS ? DE LA COLUMNA PRODUCTO


df_union = df_union.replace('Aceite, algod?n', 'Aceite, Algodón')
df_union = df_union.replace('algod?n','algodón')
df_union = df_union.replace('Aceite, c?rtamo, alazor', 'Aceite, cártamo, alazor')
df_union = df_union.replace('Aceite, ma?z','Aceite, maíz' )
df_union = df_union.replace('Almendras con c?scara', 'Almendras con cáscara')
df_union = df_union.replace('An?s, badi?n, hinojo, cilantro', 'Anís, badián, hinojo, cilantro' )
df_union = df_union.replace('Azucar, ca?a', 'Azucar, caña')
df_union = df_union.replace('B?falos', 'Búfalos')
df_union = df_union.replace('Caf?, verde', 'Café, verde')
df_union = df_union.replace('C??amo, estopa y desechos', 'Cáñamo, estopa y desechos')
df_union = df_union.replace('Ca?am?n', 'Cañamón')
df_union = df_union.replace('Casta?as', 'Castañas')
df_union = df_union.replace('Coles y otras cruc?feras', 'Coles y otras crucíferas')
df_union = df_union.replace('Coliflor y br?col', 'Coliflor y brócoli')
df_union = df_union.replace('Algod?n con semilla', 'Algodón con semilla')
df_union = df_union.replace('Algod?n, fibra', 'Algodón, fibra')
df_union = df_union.replace('Ar?ndanos trepadores', 'Arándanos trepadores')
df_union = df_union.replace('Avellanas, con c?scara', 'Avellanas, con cáscara')
df_union = df_union.replace('D?tiles', 'Dátiles')
df_union = df_union.replace('Frutas, pom?ceas nep', 'Frutas, pomáceas nep')
df_union = df_union.replace('Leche, entera fresca, b?fala', 'Leche, entera fresca, búfala')
df_union = df_union.replace('L?pulo', 'Lúpulo')
df_union = df_union.replace('Man?s (cacahuete) con c?scara', 'Manís (cacahuete) con cáscara')
df_union = df_union.replace('Nueces con c?scara', 'Nueces con cáscara')
df_union = df_union.replace('Puerros y otras ali?ceas', 'Puerros y otras aliáceas')
df_union = df_union.replace('Ra?ces de achicoria', 'Raíces de achicoria')
df_union = df_union.replace('Sand?as', 'Sandías')
df_union = df_union.replace('Semilla de algod?n', 'Semilla de algodón')
df_union = df_union.replace('Ar?ndanos', 'Arándanos')


```
